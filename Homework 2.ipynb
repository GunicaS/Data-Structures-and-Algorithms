```python
class Stack:
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if self.items:
            return self.items.pop()
        else:
            print("pop() error: Stack is empty.")
            return None

    def peek(self):
        if self.items:
            return self.items[-1]
        else:
            print("peek() error: Stack is empty.")
            return None

    def size(self):
        return len(self.items)


"""
1. postfixEval
Input type: a list of strings
Output type: a floating point number
"""
def postfixEval(expr):
    stack = Stack() #defined stack 
    for i in expr: #for each value in the input
        try: #used try and except to avoid errors 
            i = float(i) #converted the value in expression to a float
            stack.push(i) #pushed this now converted value into the float
        except: #if error happens where the value cannot be a float the following code is run 
            number1 = stack.pop() #popped off one number from stack and assigned to a variable 
            number2 = stack.pop() #popped off second number from stack and assigned to a variable 
            if i == "+": #if the value equals the addition sign the number variables are added together
                newnum = number2 + number1
                stack.push(newnum) #the new number is pushed onto stack 
            elif i == "-": #same process is repeated for all of the other operations
                newnum = number2 - number1
                stack.push(newnum)
            elif i == "*":
                newnum = number2 * number1
                stack.push(newnum)
            elif i == "/":
                newnum = number2 / number1
                stack.push(newnum)
    return stack.peek() #the last number on the stack is peeked and returned which should be the final number after all of the operations

"""
2. validParentheses
Input type: a string
Output type: a Boolean
"""
def validParentheses(s):
    stack = Stack() #stack is defined
    for i in s: #for each value in the given input
        if i in "({[": #if the value is one of the valid open brackets
            stack.push(i) #the value is pushed on the stack
        elif i in ")}]": #if the value is one of the valid closing brackets 
            if stack.isEmpty(): #check if the stack is empty
                return False #if the stack is empty False is returned 
            startbracket = stack.pop() #pop off one of the stack brackets to start comparing 
            if not ((startbracket == "(" and i == ")") or (startbracket == "{" and i == "}") or (startbracket == "[" and i == "]")): #compare the start brackets with the end brackets
                return False #if any of the start brackets do not match return false
    return stack.isEmpty() #return the stack when it is empty



"""
3. reverseString
Input type: a string
Output type: a string
"""
def reverseString(s):
    stack = Stack() #defined the stack 
    reverse = "" #defined a string which will hold the reverse values 
    for i in s: #for each value in s(input)
        stack.push(i) #the value is pushed onto the stack
    while stack.isEmpty() != True: #while the stack is not empty 
        reverse = reverse + stack.pop() #the new string is added too by popping off the stack 
    return reverse #the new string is return
```
