```python
"""
Queue class used for Problem 1
"""
class Queue:
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []

    def enqueue(self, item):
        self.items.insert(0, item)

    def dequeue(self):
        return self.items.pop()

    def size(self):
        return len(self.items)

"""
Node class used for Problem 2-5
"""
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

    def getData(self):
        return self.data

    def getNext(self):
        return self.next

    def setData(self, data):
        self.data = data

    def setNext(self, node):
        self.next = node



"""
1. Stack2 class
Implement stack data structure using queue
"""
class Stack2:
    def __init__(self):
        self.queue1= Queue() #initalized the queue 
        
    def isEmpty(self):
        return self.queue1.isEmpty() #to check if the queue is empty then check if the queue1 is empty

    def push(self, item):
        self.queue1.enqueue(item) #pushing an item onto the stack is the same as enqueueing in a stack so use the enqueue function

    def pop(self):
        newqueue = Queue() #create a new temporary queue
        if self.queue1.isEmpty(): #if the original queue is empty return None
            return None
        for i in range((self.queue1.size())-1): #since we want to pop off an item we take each item in the queue - 1 because we want that last item
            newqueue.enqueue(self.queue1.dequeue()) #queue all of the items execpt the last one onto the new queue 
        poppeditem = self.queue1.dequeue() #can pop off the last item we want
        while not newqueue.isEmpty(): #while the queue is not empty 
            self.queue1.enqueue(newqueue.dequeue()) #we can enqueue the items back onto the original queue 
        return poppeditem #return the item that was "popped off"

    def peek(self):
        newqueue = Queue() #create a new temporary queue 
        if self.queue1.isEmpty(): #if the original queue is empty return None
            return None
        for i in range((self.queue1.size())- 1): #take all of the items in the queue execpt the last one
            newqueue.enqueue(self.queue1.dequeue()) #enqueue all of these items onto the new queue
        peekeditem = self.queue1.dequeue() #dequeue the peeked item
        newqueue.enqueue(peekeditem) #enqueue the peeked item back onto the new queue so that the origial item ordered is restored
        while not newqueue.isEmpty(): #while the new queue is not empty
            self.queue1.enqueue(newqueue.dequeue()) #dequeue the items from the temporary queue back onto the new queue 
        return peekeditem #return the peeked item

    def size(self):
        return self.queue1.size() #to get the size just use the size() on the defined queue


"""
2. transform(lst)
Transform an unordered list into a Python list
Input: an (possibly empty) unordered list
Output: a Python list
"""
def transform(lst):
    newlist = [] #define the list
    currentnode = lst #the current node is equal to the lst
    while currentnode is not None: #while the current node not equal to the None 
        newlist.append(currentnode.getData()) #append the data from the node onto the list
        currentnode = currentnode.getNext() #move to the next node and repeat till the last node is reached
    return newlist #return the list



"""
3. concatenate(lst1, lst2)
Concatenate two unordered lists
Input: two (possibly empty) unordered list
Output: an unordered list
"""
def concatenate(lst1, lst2): 
    if lst1 is None: #if the first list is none 
        return lst2 #return the second list
    if lst2 is None: #if the second list is none
        return lst1 #return the first list 
    currentnode = lst1 #set the current node to the first list
    while currentnode.next is not None: #while the next node is not none 
        currentnode = currentnode.getNext() #get the next node
    currentnode.next = lst2 #set the next node as the node for the second list 
    return lst1 #return the first list which has both the lists on it concatenated



"""
4. removeNodesFromBeginning(lst, n)
Remove the first n nodes from an unordered list
Input:
    lst -- an (possibly empty) unordered list
    n -- a non-negative integer
Output: an unordred list
"""
def removeNodesFromBeginning(lst, n):
    counter = 0 #set a counter 
    while lst and counter < n: #while the list is less than the number of nodes to remove 
        lst = lst.getNext() #get the next node 
        counter = counter + 1 #iterate the counter with + 1 and the loop runs back till n is reached
    return lst #the list with the nodes removed is returned



"""
5. removeNodes(lst, i, n)
Starting from the ith node, remove the next n nodes
(not including the ith node itself).
Assume i + n <= lst.length(), i >= 0, n >= 0.
Input:
    lst -- an unrdered list
    i -- a non-negative integer
    n -- a non-negative integer
Output: an unordred list

lst = [1, 2, 3, 4, 5]
i = 2
n = 2
return [1, 2, 5]

i = 1
n = 2
return [1, 4, 5]

i = 0
n = 2
return [3, 4, 5]
"""
def removeNodes(lst, i, n):
    if lst is None or n == 0: #if the list has no values or the number of nodes to remove is equal to 0 
        return lst #the list is returned as is 
    currentnode = lst #the current node is set to the list 
    if i > 0: #if the node number is greater than 0 
        for item in range(i-1): #traverses to the node - 1
            if currentnode is None: #if the node is None
                return lst #the list is returned 
            currentnode = currentnode.getNext() #when the node is not None keep going to the next node 
    else: #if i = 0 then the just go to the next node 
        return lst.getNext()
    if currentnode is not None: #when the current node is not none 
        nextnode = currentnode.getNext() #iterate to the next node 
        for item in range(n): #for each item in the range to the number of nodes to remove 
            if nextnode is None: #if the next node is none the function breaks as all the nodes would have been removed
                break
            nextnode = nextnode.getNext() #if none is not reached the next node is iterated
        currentnode.setNext(nextnode) #the current node is set to the nextnode skipping over all the nodes we need to
    return lst #the list returned
```
