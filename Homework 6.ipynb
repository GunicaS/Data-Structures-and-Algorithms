```python

class BinaryTreeNode:
    def __init__(self, data):
        self.val = data
        self.left = None
        self.right = None



"""
1. Iterative preorder traversal of a binary tree
"""
def preorder(root):
    tempstack = [] #define two stacks one a temp stack other a result stack
    result = []

    if not root: #checks to see if the root is none
        return [] #if the root is non returns empty list
    
    while root or tempstack: #defines the while loop that runs while the root is none or tempstack is not empty
        if root: 
            result.append(root.val) #appends value of root to the result stack
            tempstack.append(root.right) #appends the right value onto the temp stack
            root = root.left #changes the root to the left child
        else:
            root = tempstack.pop() #if the root is none it pops a node from the temp stack allowing it to go down the right subtree

    return result #returns the preorder traversal 



"""
2. Reconstruct Binary Tree
"""
# inorder traversal can be divided as [left-subtree-nodes, root, right-subtree-nodes]
# preorder traversal can be divided as [root, left-subtree-nodes, right-subtree-nodes]
def reconstructBT(preorder, inorder):
    if not preorder or not inorder: #checks if the preorder and inorder lists are empty or not
        return None #if empty returns none
    
    rootnode = BinaryTreeNode(preorder[0]) #creates a new binary tree object and it assigns to it as the rootnode
    rootnodeinorder = inorder.index(preorder[0]) #finds the location of the rootnode in the inorder list

    rootnode.left = reconstructBT(preorder[1:rootnodeinorder + 1], inorder[:rootnodeinorder]) #using recursion constructs the left subtree from the current node 
    #passes the preorder from to rootnodeinorder + 1 and the inorder sublist from the rootnodeinorder

    rootnode.right = reconstructBT(preorder[rootnodeinorder + 1:], inorder[rootnodeinorder + 1:]) #using recursion constructs the right subtree from the current node
    #passes the preorder from to rootnodeinorder + 1 and the inorder sublist from the rootnodeinorder + 1 to the end 

    return rootnode #returns the current node



"""
3. Convert Binary Search Tree
"""
def convertBSTtoGST(root): #from TA office hours
    c = [0] #list with a single element is intializied 

    def helper(root, c): #a helper function with 2 parameters root and c is initalized
        if root is None: #if root is none the function returns as there are no more nodes to process
            return
        
        helper(root.right, c) #recursively calls the helper function on the right subtree of the current node

        print(root.val) #prints the value of the current node

        c[0] += root.val #value of current node is added to cummulative sum 

        root.val = c[0] #the root value is set to equal the cummulative sum

        helper(root.left,c) #recursively calls the helper function on the left subtree of the current node

    helper (root,c) #initates the helper function
    
    return root #returns the root of BST which represents the Greater Sum Tree
```
